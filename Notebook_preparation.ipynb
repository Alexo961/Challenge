```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import  pickle

%matplotlib inline
```

```python
def load_data(csv):
    dataset=pd.read_csv(csv)
    return dataset
```

**Stampo Le chiavi come lista**

```python
def chiavi(dataset):
    for i in dataset:
        print ("La colonna :",i,"ha le chiavi",dataset[i].unique())
```

```python
def X_y(dataset,target):
        y= dataset[target]
        X= dataset.drop([target],axis=1)
        return X,y
```

```python
def split(X,y):
         X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2,random_state=0)
         return  X_train, X_test, y_train, y_test
```

**Visualizzo il numero di chiavi uniche delle colonne**

```python
def visualizza(dataset):
     for i in dataset:
        print("colonna :",i,"conta", dataset[i].nunique(),"valori di chiavi diversi")
    
```

```python
def crea_csv(X,y):
    X_train,X_test,y_train,y_test =split(X,y)
    
    
    output_train = pd.DataFrame(pd.concat([X_train,y_train],axis = 1))
    
    
    output_test =  pd.DataFrame(pd.concat([X_test,y_test],axis = 1))
   
    return output_train,output_test
```

```python
def crea_file(output_train,output_test):
    output_train.to_csv('train.csv', index=False)
    output_test.to_csv('test.csv', index=False)
    
```
